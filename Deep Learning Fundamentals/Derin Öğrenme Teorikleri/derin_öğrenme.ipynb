# FORWARD - BACK PROPOGATİON

1. Forward Propagation (İleri Yayılım)

Giriş verileri nörona gelir.
Her giriş, ağırlığıyla çarpılır ve bias eklenir.
Bu toplam aktivasyon fonksiyonuna sokularak bir çıktı elde edilir.
Bu süreç, ağdaki tüm katmanlar boyunca tekrar eder ve en sonunda bir tahmin yapılır.

2. Loss Hesaplama (Hata Bulma)

Modelin ürettiği tahmin ile gerçek değer karşılaştırılır.
Aradaki fark, yani hata hesaplanır.

3. Backpropagation (Geri Yayılım)

Hatanın sebebini anlamak için geriye doğru hesaplamalar yapılır.
Hata, hangi ağırlıklardan kaynaklanıyor diye zincirleme türev yöntemiyle hesaplanır.
Ağırlıklar, hatayı minimize edecek şekilde güncellenir.

### 1. Precision (Kesinlik):

Precision, modelin pozitif sınıfı doğru tahmin etme oranıdır.
Başka bir deyişle, modelin pozitif olarak tahmin ettiği örneklerin ne kadarının gerçekten pozitif olduğunu ölçer.
Yüksek Precision değeri, modelin pozitif tahminlerinde hata yapma olasılığının düşük olduğunu gösterir.

Precision= True Positives/(True Positives+False Positives)


True Positives (TP): Doğru bir şekilde pozitif olarak sınıflandırılan örnekler.

False Positives (FP): Hatalı bir şekilde pozitif olarak sınıflandırılan örnekler.

### 2. Recall (Duyarlılık veya Gerçek Pozitif Oranı):

Recall, modelin gerçek pozitif örnekleri doğru tespit etme oranıdır.

Başka bir deyişle, modelin gerçek pozitifleri ne kadar doğru bulduğunu ölçer.
Yüksek Recall değeri, modelin gerçek pozitifleri kaçırma oranının düşük olduğunu gösterir.

Recall= (True Positives)/ (False Negatives+True Positives)
​
 
True Positives (TP): Doğru pozitif tahminler.
False Negatives (FN): Gerçek pozitifler olup modelin yanlışlıkla negatif olarak sınıflandırdığı örnekler.

### 3. F1 Skoru:

F1 Skoru, Precision ve Recall'un harmonik ortalamasıdır. Yani, her ikisinin de iyi olması gereken durumlarda F1 skoru önemli bir değerlendirme metrikidir.

F1 Skoru, genellikle dengesiz veri setlerinde (örneğin, pozitif sınıf çok azsa) iyi bir değerlendirme metrikidir çünkü her iki ölçütü de dengelemeye çalışır.

Precision ve Recall arasında bir denge sağlamaya çalışır. Bu nedenle, hem Precision hem de Recall önemli olduğunda F1 Skoru kullanılır.

#### Özet:

Precision: Modelin pozitif sınıflandırmalarının ne kadar doğru olduğunu ölçer.
Recall: Modelin gerçek pozitifleri ne kadar iyi bulduğunu ölçer.
F1 Skoru: Precision ve Recall arasındaki dengeyi ölçer ve her iki metrik de önemli olduğunda kullanılır.

# GENEL ÖZET

## 1. Katman Sayısı ve Nöron Sayısı

Sinir ağındaki her katman, belirli bir görev için özelleşmiştir. Genelde giriş katmanı (input layer), gizli katmanlar (hidden layers) ve çıkış katmanı (output layer) olmak üzere üç ana kısımdan oluşur.


* Giriş Katmanı: Girdi verilerinin doğrudan ağ modeline sunulduğu ilk katmandır.
 Bu katmanın nöron sayısı, girdi veri setinin özelliklerine göre belirlenir. Örneğin, bir resim verisi için her bir piksel bir nöron olur, yani resmin boyutlarına göre belirlenir.


* Gizli Katmanlar: Giriş katmanından çıkan veriler, çeşitli dönüşümlere uğrayarak modelin öğrendiği temsillere dönüşür. Gizli katmanlar, ağın "öğrenme kapasitesini" belirler. Gizli katmanlardaki nöron sayısı, genellikle hiperparametre olarak belirlenir ve bu sayı arttıkça modelin kapasitesi (ve karmaşıklığı) artar.

* Çıkış Katmanı: Sonuçları üreten katmandır. Çıkış katmanındaki nöron sayısı, çözülmeye çalışılan probleme göre değişir. Örneğin, ikili sınıflandırma için 1 nöron, çok sınıflı sınıflandırma için her sınıf için bir nöron bulunur.

#### Ekstra Not: 

Derin öğrenme modelleri daha derin katmanlar kullanarak daha karmaşık özellikleri öğrenebilir, ancak aşırı derin ağlar vanishing gradient ya da exploding gradient gibi problemlere yol açabilir. Bu yüzden derinlik arttıkça uygun aktivasyon fonksiyonları seçilmelidir (örneğin, ReLU).

## 2. Ağırlıklar Nasıl Hesaplanır?

Ağırlıklar, nöronlar arasındaki bağlantıları tanımlar ve her bağlantının "önemini" belirler. Bu ağırlıklar genellikle başlangıçta rastgele atanır ve model eğitildikçe öğrenilir. Ağırlıkların hesaplanmasında kullanılan temel adımlar şunlardır:

* Başlangıç Ağırlıkları: Ağırlıklar genellikle rastgele seçilir. Bu rastgelelik, modelin yerel minimumlardan kaçmasına yardımcı olur ve daha genel çözümler bulmasını sağlar. Ancak, çok büyük ya da çok küçük ağırlıklar başlangıçta aşırı eğitim (overfitting) veya öğrenme sürecinde sorunlara yol açabilir.

* Aktivasyonlar ve İleri Yayılım: Bir katmanda bir nöronun çıktısını bulmak için, girdi verilerini, ağırlıkları ve bias (sapma) terimini çarparak toplarız ve bunu bir aktivasyon fonksiyonu ile işleriz.

#### Ekstra Not:
 Başlangıç aşamasında kullanılan optimizasyon tekniklerinin doğru seçimi de önemlidir. Hedef modelin doğruluğu değil, en hızlı ve verimli şekilde öğrenmesini sağlamak olmalıdır.



## 3. AĞIRLIKLAR NEYE GÖRE SEÇİLİR ?

* Ağırlık Başlangıcı:

Ağırlıklar başlangıçta rastgele seçilir, çünkü sabit bir değerle başlamak modelin öğrenmesini zorlaştırabilir. Başlangıç için kullanılan bazı popüler yöntemler:

Xavier ve He gibi yöntemler, ağırlıkları önceki katmandaki nöron sayısına göre ayarlar. Bu, modelin öğrenmesini hızlandırır.

---- Xavier, genelde sigmoid ve tanh aktivasyonları için uygundur.
He, ReLU gibi aktivasyonlar için daha iyi sonuçlar verir.

* Ağırlık Güncellenmesi (Optimizasyon):

Model eğitildikçe, ağırlıklar optimizer fonksiyonlarıyla güncellenir. Yani, modelin doğru sonuç vermesi için ağırlıklar sürekli olarak ayarlanır.

1. Gradient Descent (Gradyan İnişi):

En temel optimizasyon yöntemidir. Hata (kaybı) minimize etmek için ağırlıklar, hata gradyanına göre güncellenir. Hedef, her seferinde ağırlıkları biraz daha doğru hale getirmektir.

2. Adam (Adaptive Moment Estimation):

Adam, öğrenme hızını her parametreye uyarlayarak daha hızlı ve stabil bir öğrenme sağlar. Momentum kullanarak geçmişteki gradyanları dikkate alır, bu da ağırlıkların daha doğru bir şekilde güncellenmesini sağlar.

3. RMSprop:

Bu, Adam'a benzer şekilde, ağırlıkları uyarlamalı öğrenme hızlarıyla günceller. Ancak, sadece gradyanın karesine bakar, bu da daha hızlı ve kararlı bir öğrenme sağlar.

#### Ekstra Not: 
Leaky ReLU ve Parametric ReLU gibi türevleri de vardır. Bunlar, özellikle ReLU'nun "ölü nöron" (dead neurons) problemini çözer.

# 4. KATMAN SAYILARI NEYE GÖRE BELİRLENİR

Katman sayısını ve her katmandaki nöron sayısını belirlemek için genellikle bazı temel yöntemler ve düşünceler kullanılır.

1. Problem Karmaşıklığı:

Basit problemler için daha az katman ve daha az nöron yeterli olabilir. Örneğin, doğrusal ilişkiler veya basit sınıflandırma problemleri için 1-2 katmanlı bir model yeterli olabilir.

Karmaşık problemler (örneğin, görüntü sınıflandırma, doğal dil işleme) için daha derin ağlar (daha fazla katman ve nöron) gerekir. Derin ağlar, verilerdeki karmaşık ilişkileri öğrenme yeteneğine sahiptir.

2. Veri Miktarı:

Yeterli veri varsa, daha derin bir ağ (daha fazla katman ve nöron) daha iyi sonuç verebilir. Ancak, veriniz sınırlıysa, çok derin ağlar aşırı öğrenmeye (overfitting) neden olabilir.

Az veri ile çok derin ağlar kullanmak overfitting'e yol açabilir. Bu durumda daha yüzeysel (daha az katmanlı) modeller tercih edilebilir.

3. Katman Sayısını ve Nöronları Belirlemek İçin Kullanılan Yöntemler:

* Deneysel Yöntemler (Empirical Methods):

En yaygın yöntem, farklı katman sayıları ve nöron sayılarını deneyerek en iyi sonuçları veren yapıyı bulmaktır. Bu işlem, grid search veya random search gibi optimizasyon teknikleriyle yapılabilir.

Cross-validation yöntemiyle modellerin doğruluğu test edilir ve en uygun yapı seçilir.

* Heuristik Yöntemler:

Gizli katmanların sayısı genellikle veri setinin karmaşıklığına göre belirlenir. Görüntü işleme gibi karmaşık görevler için daha fazla katman gerekir.
Katman başına nöron sayısı genellikle deneysel olarak belirlenir ama başlangıçta şu kurallar kullanılabilir:

-- İlk katmanlar daha geniş (daha fazla nöron), son katmanlar ise daha dar olabilir.
Örneğin, giriş katmanındaki nöron sayısı, giriş verisinin boyutuna göre belirlenir (örneğin, bir görüntüde 28x28 piksellik giriş için 784 nöron).

 4. Hyperparameter Tuning:
 
Katman ve nöron sayıları, ağın hiperparametreleri arasında yer alır. Bu hiperparametreler genellikle grid search, random search, bayesian optimization gibi yöntemlerle optimize edilir.

#### Ekstra Not: 
Huber loss gibi hybrid kayıp fonksiyonları da mevcuttur, bu fonksiyonlar her iki kayıp fonksiyonunun avantajlarını birleştirir.

# 5. Aktivasyon Fonksiyonu Seçimi:
Aktivasyon fonksiyonu, nöronun çıktısını belirler ve genellikle problem türüne göre seçilir. İşte bazı yaygın kullanım örnekleri:

#### ReLU (Rectified Linear Unit):

* Kullanım Alanı: Derin sinir ağları (DNN), konvolüsyonel sinir ağları (CNN) ve genellikle çok katmanlı yapılar için kullanılır.

* Neden Seçilir?: ReLU, negatif girişleri sıfırlar, pozitifleri olduğu gibi bırakır. Bu, modelin daha hızlı öğrenmesine ve gradyan kaybolma (vanishing gradient) problemini azaltmasına yardımcı olur. Genellikle gizli katmanlarda tercih edilir.

#### Sigmoid:

* Kullanım Alanı: Binary classification (ikili sınıflandırma) için çıkış katmanında yaygındır.

* Neden Seçilir?: Çıktıyı 0 ile 1 arasında sınırladığı için, bir sınıfın olasılığını temsil etmek için idealdir. Ancak, gradyan kaybolma sorununa yol açabilir ve genellikle derin ağlarda tercih edilmez.

#### Tanh (Hyperbolic Tangent):

* Kullanım Alanı: Sınıflandırma ve regresyon gibi genel görevlerde kullanılabilir, genellikle derin ağlarda gizli katmanlarda tercih edilir.

* Neden Seçilir?: Tanh, çıkışı -1 ile 1 arasında sınırlar, bu da daha dengeli bir çıktı sağlar. Ancak, gradyan kaybolma yine de bir sorun olabilir.

#### Softmax:

* Kullanım Alanı: Multi-class classification (çok sınıflı sınıflandırma) problemleri için çıkış katmanında kullanılır.

* Neden Seçilir?: Çıkışı olasılık dağılımı olarak verir ve her bir sınıfın olasılığını hesaplar. Çoklu sınıflar arasında en yüksek olasılığı seçerek tahmin yapar.

# 6. Kayıp Fonksiyonu Seçimi:

Kayıp fonksiyonu (loss function), modelin ne kadar hatalı olduğunu hesaplar ve modelin öğrenmesini yönlendirir. Seçimi de problemin türüne bağlıdır:

#### Mean Squared Error (MSE):

* Kullanım Alanı: Regresyon problemleri için kullanılır.

* Neden Seçilir?: Gerçek değerlerle tahmin edilen değerler arasındaki farkların karesinin ortalamasını alır. Bu, sürekli değişkenler için uygundur.

#### Binary Cross-Entropy (Log Loss):

* Kullanım Alanı: İkili sınıflandırma problemleri için kullanılır.

* Neden Seçilir?: İkili sınıflandırmada, modelin her sınıf için olasılığını hesaplar ve doğru sınıf için logaritmik hata hesaplar.

#### Categorical Cross-Entropy:

* Kullanım Alanı: Çok sınıflı sınıflandırma problemleri için kullanılır.

* Neden Seçilir?: Birden fazla sınıf için doğruluk ölçümü yapar ve modelin sınıf tahminlerinin doğruluğunu hesaplar.

# 7. Optimizasyon Algoritması Seçimi:
Optimizasyon algoritması, modelin öğrenme hızını ve doğruluğunu etkileyen önemli bir faktördür. Yaygın kullanılan optimizasyon algoritmaları:

#### SGD (Stochastic Gradient Descent):

* Kullanım Alanı: Genel amaçlı kullanılır.

* Neden Seçilir?: Temel gradient descent algoritmasıdır ve küçük adımlar atarak yerel minimuma inmeye çalışır. Ancak momentum veya learning rate gibi hiperparametre ayarları ile optimize edilmesi gerekir.

#### Adam (Adaptive Moment Estimation):

* Kullanım Alanı: Çoğu derin öğrenme problemleri için yaygın olarak kullanılır.

* Neden Seçilir?: Hem öğrenme oranını hem de momentumu otomatik olarak adapte eder. Bu, genellikle daha hızlı öğrenmeye ve daha iyi sonuçlar elde etmeye yardımcı olur.

#### RMSProp:

* Kullanım Alanı: Derin ağlar ve LSTM gibi model türleri için kullanılır.

* Neden Seçilir?: Düşük öğrenme oranlarıyla daha kararlı bir optimizasyon sağlar ve genellikle daha hızlı bir öğrenme süreci sağlar.

#### Ekstra Not:
 AdaGrad gibi algoritmalar da daha seyrek güncellenen parametrelerde (örneğin, metin verisi üzerinde çalışırken) daha verimli olabilir.

# 8. Hiperparametre Optimizasyonu:
Farklı fonksiyonları ve yapıların seçimleri için deneysel yöntemler kullanabilirsiniz. Hiperparametrelerin belirlenmesinde kullanılan yöntemler:

* Grid Search: Birçok farklı hiperparametre kombinasyonu ile denemeler yaparak en iyi sonucu elde etmeye çalışır.

* Random Search: Hiperparametre alanında rastgele seçimler yaparak potansiyel en iyi kombinasyonu bulur.

* Bayesian Optimization: Hiperparametreleri daha verimli bir şekilde optimize etmek için kullanılır.

#### Ekstra Not: 
Son zamanlarda, genetik algoritmalar veya differential evolution gibi daha gelişmiş arama yöntemleri de popüler hale gelmiştir.

### 2. Başlangıçta rastgele seçilen ağırlıkların amacı nedir ve hangi sorunları engellemeye yardımcı olur?
Başlangıçta ağırlıkları rastgele seçmek, modelin farklı özellikleri öğrenmesine olanak tanır ve daha genel çözümler bulunmasına yardımcı olur. Bu, modelin öğrenme sürecinde çeşitlilik sağlar.
Sıfır ağırlıklar, simetrik ağırlık güncellemesi problemini doğurur ve modelin evrimleşmesini engeller.

### 3. Peki, Xavier ve He gibi yöntemler neden önemlidir ve hangi aktivasyon fonksiyonlarıyla daha iyi sonuç verir?

* He Initialization: Bu yöntem, özellikle ReLU (Rectified Linear Unit) aktivasyon fonksiyonuyla kullanılan ağırlık başlatma yöntemidir. He, ağırlıkları normal dağılım ile başlatır, ancak standart sapmayı ağın önceki katmanındaki nöron sayısına göre ayarlar. Bu, ReLU'nun negatif değerleri sıfıra dönüştürme özelliğinden dolayı daha iyi performans sağlar. ReLU'nun gradyan kaybolma (vanishing gradient) problemine karşı daha dayanıklı olmasını sağlar ve modelin daha hızlı öğrenmesine yardımcı olur.

* Xavier Initialization: Bu yöntem ise genellikle sigmoid veya tanh gibi aktivasyon fonksiyonlarıyla daha uyumludur. Xavier, ağırlıkları uniform (düzgün) dağılım ile başlatır ve önceki katmandaki nöron sayısının karesine bağlı olarak ağırlıkların büyüklüğünü ayarlar. Bu, aktivasyon fonksiyonlarının daha dengeli çalışmasını sağlar ve modelin daha verimli öğrenmesini sağlar. Xavier, genellikle sigmoid ve tanh gibi sınırlı çıkış aralıklarına sahip fonksiyonlarla iyi çalışır.

* Kısacası:

He başlatması: ReLU için

Xavier başlatması: sigmoid ve tanh için

### 4. Peki, ağırlık güncellemeleri sırasında kullandığımız optimizasyon algoritmalarından hangisinin avantajları daha fazla? Örneğin, Adam ve SGD arasındaki fark nedir?
* Adam, genellikle daha hızlı ve stabil sonuçlar sağlarken, SGD daha temel bir yöntem olup doğru hiperparametre ayarları ile etkin olabilir. Ancak, derin öğrenme projelerinde ve büyük veri setlerinde genellikle Adam tercih edilir.

* SGD sabit bir öğrenme hızı kullanırken, Adam her parametre için uyarlanabilir öğrenme hızı kullanır.
Adam, gradyanların geçmiş momentumu ve 2. momenti (yani, gradyanların karelerinin ortalaması) kullanarak daha stabil bir öğrenme sağlar.
SGD genellikle daha yavaş öğrenme eğilimindeyken, Adam daha hızlı ve stabil bir optimizasyon sağlar.

### 5. Şimdi, optimizer konusu açıklığa kavuştu. Hangi aktivasyon fonksiyonları ne tür sorunlar için daha uygundur?

* ReLU: Derin ağlar için yaygın olarak kullanılan bir aktivasyon fonksiyonudur. Genellikle gizli katmanlarda kullanılır ve regresyon problemleri için uygundur.
Sigmoid: İkili sınıflandırma problemleri için ideal olup, çıkış katmanında tercih edilir.

* Tanh: Gizli katmanlarda negatif ve pozitif değerlerin simetrik olmasına yardımcı olur, genellikle regresyon veya sınıflandırma problemlerinde kullanılır.

* Softmax: Çoklu sınıflı sınıflandırma problemleri için çıkış katmanında kullanılır ve her sınıf için olasılık tahmin eder.

### 6. Kayıp Fonksiyonu (Loss Function) Türlerini Yazınız

```python
# Gerekli Kütüphaneleri İçe Aktarma
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.activations import relu, sigmoid

# Veri setini hazırlama (örnek veri, burayı kendi verinizle değiştirin)
# X: Özellikler (features), Y: Etiketler (labels)
X_train = np.random.rand(1000, 784)  # Örnek veri: 1000 örnek, 28x28 boyutunda resimler (784 özellik)
Y_train = np.random.randint(2, size=1000)  # 0 ve 1 arasında etiketler (ikili sınıflandırma)

# Modeli Tanımlama
model = Sequential()

# Giriş Katmanı (Input Layer) - 784 nöron, her bir piksel bir nöron
model.add(Dense(512, input_dim=784, activation=relu))  # Gizli katman: 512 nöron, ReLU aktivasyon
model.add(Dropout(0.2))  # Dropout eklenmesi, aşırı öğrenmeyi (overfitting) engellemek için

# Gizli Katman (Hidden Layer) - 256 nöron, ReLU aktivasyonu
model.add(Dense(256, activation=relu))
model.add(Dropout(0.2))

# Çıkış Katmanı (Output Layer) - 1 nöron, sigmoid aktivasyonu (ikili sınıflandırma)
model.add(Dense(1, activation=sigmoid))

# Modelin Derlenmesi (Compile)
model.compile(optimizer=Adam(learning_rate=0.001), 
              loss=BinaryCrossentropy(), 
              metrics=['accuracy'])

# Modeli Eğitme
model.fit(X_train, Y_train, epochs=10, batch_size=32, validation_split=0.2)

# Modeli Değerlendirme
loss, accuracy = model.evaluate(X_train, Y_train)
print(f"Model kaybı: {loss}, Model doğruluğu: {accuracy}")
```
